# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([5,6,7,8])
y = np.array([3,5,6,3])
```

## Simple Plotting

### Plotting with line

```python
plt.plot(x,y) 
```

```python
plt.plot(x,y,marker="o",mfc = "black", ms = 20,mec = "orange",color = "red",linestyle = "dashed",linewidth = 10 )
```

- mfc = marker face colour
- mec = marker edge colour
- ms = marker size
- color/c = line colour
- linestyle = dotted / dashed / empty....


```python
plt.plot(x,y,"*"":b",ms = 20) # Shortcut
```

```python
plt.plot(x,y,"*""--b",ms = 20) # Shortcut
```

### Plotting without line

```python
plt.plot(x,y,marker="o",mfc = "black", ms = 20,mec = "orange",color = "red",linestyle = "",linewidth = 10 )
```

```python
plt.plot(x,y,"*""k",ms = 20) # Shortcut
```

## Scatter plot

```python
colors = ['red','blue','orange','black']
size = [100,200,150,250]

plt.scatter(x,y,color = colors, s = size, marker = "*") # basically ("*", color = ... ) of plt.plot but much easier (no joining lines)
plt.title("haha",loc = "left")                         # ms doesn't work here, s,c and color work
plt.ylabel("hello")
plt.show
```

- All points of scatter can be set to same size as well as same colour by giving a single value to appropriate arguments rather than passing arrays.
- For a normal plot, we cannot have different sizes and colours for different points, but it can be done in case of scatter plot.

## Bar graphs

```python
colors = ['r','g','b','k']
plt.bar(x,y,color = colors, width = 0.1)
```

```python
colors = ['r','g','b','k']
plt.barh(x,y,color = colors, height = 0.5,alpha = 1) # alpha is visibility, 0 == invisible.
```

```python
colors = ['r','g','b','k']
bets = ['a','b','c','d']  # can assign strings to the axis as well.
plt.barh(bets,y,color = colors, height = 0.5,alpha = 1)
```

## Histogram

```python
z = np.random.normal(170,100,500)
plt.hist(z) # since only one parameter is passed, the parameter y is as a default set to frequency.
plt.show()
z   
```

## Pie Charts

```python
color = ["red","blue","yellow","green"]
labels = ['h','a','g','y']
explode = [1,2,3,0]       
plt.pie(y,labels = labels, explode = explode,startangle = 90, colors = color)
plt.legend(labels = labels)
plt.show()
```

- labels: shows what colour represents each arc
- explode: shows how far each arc is from a regular pie chart
- startangle: determined the starting angle of first data


## Labelling and Grids

```python
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("x",fontsize = 20, color = "red", loc = "left")
plt.ylabel("y",fontsize = 20, color = "blue", loc = "bottom")
plt.grid()
```

```python
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("x",fontsize = 20, color = "red", loc = "left")
plt.ylabel("y",fontsize = 20, color = "blue", loc = "bottom")
plt.grid(axis = 'x',linestyle = "dotted",linewidth = 5) # grid limited to a single axis
```

```python
u = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
v = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(u, v, color = 'hotpink', alpha = 0)         # alpha determines the transparency of markers 

u = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
v = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(u, v, color = 'green')

plt.show()
```

## Ways of setting colours to individual points:

### By assigning a list of colours to the argument c or color (completed)

### By assigning a single colour to the argument color or c, all points will get the same colour (completed)

### RGBA format:

- Red,Green,Blue,Alpha 
- A list containing 4 values in between 0 and 1 get passed to either the argument colour or c
- The first 3 values correspond to the weights of corresponiding colurs, the last value correspond to visibliy (trasnparency of the points)

```python
plt.scatter([1,2,3],[3,2,1],c = [[0.9,0.1,0.1,0.5]]) #list of list # notice how the colours of the points are red-like
```

- If the input is given to c, the input to should be of the form "list of list"
- Otherwise, the list needs to be given to the argument color
- This is because, if the number of points match with 4, the compiler may not see the list as rgba, but corresponding colours for each point.

### Cmaps

```python
map = np.array([0, 10, 20, 30, 40, 45, 50, 55,55,5,60, 70, 80, 90, 100])
plt.scatter(u,v,c = map,cmap = "viridis")
plt.colorbar(label = 'viridis')
```

```python
map = np.array([0, 10, 20, 30, 40, 45, 50, 55,55,5,60, 70, 80, 90, 1000])
plt.scatter(u,v,c = map,cmap = "viridis_r")           # mapname_r reverses the colour map.
plt.colorbar(label = 'viridis_r')
```
