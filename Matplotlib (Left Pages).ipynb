```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]
y = [2,3,4]    # x and y need not be arrays
plt.plot(x,y)
plt.show()
```

```python
x = np.array([1,2,3])
y = np.array([3,4,5])
z = np.array([2,5,6])

plt.subplot(1,3,1)
plt.plot(x)
plt.title('1')

plt.subplot(1,3,2)
plt.plot(y)
plt.title('2')

plt.subplot(1,3,3)
plt.plot(z)
plt.title('3')
```

```python
x = [2,3,4,53,6,4,6,44,78,56]                  # scatter requires both the positional arguments.
y = [2,3,4,53,6,4,6,44,78,56] 
colors = np.array([0, 10, 20, 30, 55, 60, 70, 80, 90, 100])
plt.scatter(x,y,c = colors, cmap = "viridis")
```

```python
x = [2,3,4,53,6,4,6,44,78,56]                            # scatter requires both the positional arguments.
y = [2,3,4,53,6,4,6,44,78,56]                            # not working
colors = np.array([0, 10, 20, 30, 55, 60, 70, 80, 90, 100])
plt.bar(x,y,color = colors. cmap = "viridis")
```
