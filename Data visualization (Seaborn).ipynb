## Inferences from plots

### Distplot

```python
sns.distplot(df.Age)
```

### Relplots


```python
sns.relplot(x = 'Age', y = 'Miles', data = df, color = 'red')  # default kind = scatter
```

```python
sns.relplot(x = 'Age', y = 'Miles', data = df, color = 'red',kind = 'line') 
```

```python
sns.relplot(x = 'Age', y = 'Miles', data = df, hue = 'Product')
```

- The first plot is not different from a scatter plot except for the fact that the axes are already laabelled.
- The second plot shows that the option hue can be used to see the change with regards to other variables as well, which can't be done in scatter plot.

```python
sns.relplot(x = 'Age', y = 'MaritalStatus', data = df, hue = 'Gender') # Can use objects as variables as well.
```

### Catplot

```python
sns.catplot(x = 'Age', y = 'Miles', data = df, color = 'red',hue = 'Product') 
```

```python
sns.catplot(x = 'Product', y = 'Usage', data = df,kind = 'swarm',hue = 'MaritalStatus') 
```

```python
sns.catplot(x = 'Product', y = 'Usage', data = df,kind = 'bar',hue = 'MaritalStatus') 
```

```python
sns.catplot(x = 'Product', y = 'Usage', data = df,kind = 'violin',hue = 'MaritalStatus') 
```

```python
sns.catplot(x = 'Product', y = 'Usage', data = df,kind = 'box',hue = 'MaritalStatus') 
```

Difference between relplot and catplot:
- More number of kinds that catplot offers.
- More fine spacing that the x axis of cat plot has.
